Parsing the XML and storing the Traffic attributes in a dataframe

```
def parse_request_data(request_data):

  with open('traffic_data.xml','w') as f:
      f.write(request_data)
  xtree = et.parse("traffic_data.xml")
  xroot = xtree.getroot()

  for tags in xroot.iter('flowSegmentData'):
    xroot1=et.Element('xroot')
    xroot1=tags
    for coordinates in xroot1.iter('coordinates'):
      xroot2=et.Element('xroot')
      xroot2=(coordinates)
      for coordinate in xroot2.iter('coordinate'):
        xroot3=et.Element('xroot')
        xroot3=(coordinate)
        frc.append(xroot.find("frc").text)
        currentSpeed.append(xroot.find("currentSpeed").text)
        freeFlowSpeed.append(xroot.find("freeFlowSpeed").text)
        currentTravelTime.append(xroot.find("currentTravelTime").text)
        freeFlowTravelTime.append(xroot.find("freeFlowTravelTime").text)
        roadClosure.append(xroot.find("roadClosure").text)
        latitudes.append(coordinate.find("latitude").text)
        longitudes.append(coordinate.find("longitude").text)

df = pd.DataFrame({'Timestamp':timestamp,'FRC':frc,'Current_Speed':currentSpeed,'Free_Flow_Speed':freeFlowSpeed,'Current_Travel_Time':currentTravelTime,'Free_Flow_Travel_Time':freeFlowTravelTime,'Road_Closure':roadClosure
                  ,'Latitude':latitudes,'Longitude':longitudes})

df
```
